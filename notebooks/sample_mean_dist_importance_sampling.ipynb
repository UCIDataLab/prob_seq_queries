```python
import matplotlib.pyplot as plt
import seaborn as sns
import os
import numpy as np
```

```python
os.chdir("../")
from seq_queries.utils import read_pkl, write_pkl
```

```python
data = read_pkl("data/importance_sampling/shakespeare/mc_importance_estimate_a_rt_thou?_1000s_1024m.pkl")
GT = 0.548533
sns.kdeplot(data)
plt.axvline(np.mean(data), label = f"Mean = {round(np.mean(data),4)}", color = "green")
plt.axvline(GT, label = f"Ground Truth = 0.549", color = "red")
plt.xlabel("Marginal probability of <space>")
plt.legend()
plt.title("""Seq (hist|seq|pred|extra): 'how old a|rt|<space>|thou?'
Number of sample means: 1024
Samples per mean: 1000""")
```
