# Variance comparison

```python
GT = 0.54833
samples = [(1000,1000,1000),(5000,5000,1000),(10000,50000,1000)]
files = [["data/importance_sampling/shakespeare/mc_importance_estimate_a_rt_thou?_{}s_{}m.pkl",
          "data/importance_sampling/shakespeare/mc_importance_estimate_a_rt_thou?_{}s_{}m.pkl",
          "data/importance_sampling/shakespeare/mc_importance_estimate_a_rt_thou?_{}s_{}m.pkl"],
         ["data/random_sampling/shakespeare/mc_random_estimate_a_rt_thou?_{}s_{}m.pkl",
          "data/random_sampling/shakespeare/mc_random_estimate_a_rt_thou?_{}s_{}m.pkl",
          "data/random_sampling/shakespeare/mc_random_estimate_a_rt_thou?_{}s_{}m.pkl"]]

fig, axs = plt.subplots(3,2, figsize = (12,15))
files = list(zip(*files))

for i in range(len(samples)):
    
    data_importance = read_pkl(files[i][0].format(samples[i][0],samples[i][2]))
    data_random = read_pkl(files[i][1].format(samples[i][1],samples[i][2]))
    axs[i][0].hist(data_importance, bins = 100, color="gray")
    axs[i][1].hist(data_random, bins = 100, color = "gray")
    
    axs[i][0].axvline(np.mean(data_importance), label = f"Mean = {round(np.mean(data_importance),5)}", color = "blue")
    axs[i][0].axvline(GT, label = f"Ground Truth = 0.549", color = "red") 
    axs[i][1].axvline(np.mean(data_random), label = f"Mean = {round(np.mean(data_random),4)}", color = "blue")
    axs[i][1].axvline(GT, label = f"Ground Truth = 0.549", color = "red") 
    axs[i][0].legend()
    axs[i][1].legend()
    
    axs[i][0].set_title(f"""Number of sample means | samples per mean: {samples[i][0]} | {samples[i][2]}
Sample type: mc_importance | Variance %: {round(np.var(data_importance)*100,4)}""")
    axs[i][1].set_title(f"""Number of sample means | samples per mean: {samples[i][1]} | {samples[i][2]}
Sample type: mc_random | Variance %: {round(np.var(data_random)*100,4)}""")
    
fig.suptitle("Seq (hist|seq|pred|extra): 'how old a|rt|<space>|thou?'\n")
fig.supxlabel("Marginal probability of <space>")
fig.supylabel("Density")
fig.tight_layout()
# plt.savefig("random_importance_coverage.png")
```
