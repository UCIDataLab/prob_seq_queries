```python
%load_ext autoreload
%autoreload 2
import matplotlib.pyplot as plt
import seaborn as sns
import os
import pandas as pd
import sys
import numpy as np
import glob
import torch

sys.path.insert(1,"/home/showalte/research/prob_seq_queries/")
from seq_queries.utils import read_pkl, write_pkl
```

```python
data_path = "../data/relative_variance/shakespeare/val-dl_rel-var_reduction_{}h_{}s.pkl"
lens = [(18,20)]
```

```python
def rel_variance_reduction_oracle(data_path):
    plt.figure(figsize=(9,6))
    data = read_pkl(data_path)
    
    steps = data.shape[0]
    y = data.mean(dim=-1)
    x =range(steps)
    plt.plot(x,y)
    std = data.std(dim=-1)
    plt.fill_between(x, (y-std), (y+std), color='blue', alpha=0.1)
    plt.xlim(-0.5,60)
    plt.xlabel("Number of (sorted) beams removed")
    plt.ylabel("Relative reduction in variance of $p(X_k=a,X_{1:k} = \hat{x}|H_T)$")
    plt.title("Relative variance reduction via oracle beam removal")
    
    plt.show()
    
```

```python
data_file = data_path.format(*lens[0])
rel_variance_reduction_oracle(data_file)
```
